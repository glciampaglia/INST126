# Solutions

## Task A

```python
import pandas as pd

### BEGIN SOLUTION
def primarygen(genres):
    return genres.split(",")[0]

def load_data(fn):
    df = pd.read_csv(fn, index_col=0)
    df['Primary Genre'] = df['Genre'].apply(primarygen)
    return df
### END SOLUTION

df = load_data("data/IMDB-Movie-Data.csv")
df
```

## Task B

```python
### BEGIN SOLUTION
def summarize_genres(df):
    return df.groupby('Primary Genre').agg(
        med_revenue_mil=('Revenue (Millions)', 'median'),
        avg_rating=('Rating', 'mean'),
        num_movies=('Primary Genre', 'count')
    ).round(2)
### END SOLUTION

genre_df = summarize_genres(df)
genre_df
```

## Task C

```python
### BEGIN SOLUTON
def find_successful_genres(genre_df, minrating, minrevenue):
    return genre_df[(genre_df['avg_rating'] > minrating) & (genre_df['med_revenue_mil'] > minrevenue)]

final_df = find_successful_genres(genre_df, minrating=6.5, minrevenue=50)
final_df
### END SOLUTION
```
