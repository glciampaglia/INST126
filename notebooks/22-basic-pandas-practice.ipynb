# Solutions

## Task A

```python
import pandas as pd

### BEGIN SOLUTION
def createdf(Student, SAT, GPA, Interest, High_School_Quality):
    data = {
        'SAT': SAT,
        'GPA': GPA,
        'Interest': Interest,
        'High School Quality': High_School_Quality
    }
    df = pd.DataFrame(data, index=Student)
    return df
### END SOLUTION

createdf(Student=student20, SAT=sat20, GPA=gpa20, Interest=interest20, 
         High_School_Quality=high_school_quality20)
```

## Task B

```python
### BEGIN SOLUTION
def compute_score_col(df):
    df['Score'] = \
          (0.4 * df["GPA"] * 2) \
        + (0.3 * df["SAT"] / 160) \
        + (0.2 * df["High School Quality"]) \
        + (0.1 * df["Interest"])
    return df
### END SOLUTION

# This will test your function with a smaller data frame
test_df = pd.DataFrame.from_dict({
    'Abbess Horror': {
        'SAT': 1300, 'GPA': 3.61, 'Interest': 10, 'High School Quality': 7
    },
    'Adele Hawthorne': {
        'SAT': 1400, 'GPA': 3.67, 'Interest': 0, 'High School Quality': 9
    },
    'Adelicia von Krupp': {
        'SAT': 900, 'GPA': 4.0, 'Interest': 5, 'High School Quality': 2
    }
}, orient="index")

compute_score_col(test_df)
```

## Task C

```python
### BEGIN SOLUTION
def compute_outlier_col(df):
    df['Is_outlier'] = (df['Interest'] == 0) | ((df['GPA'] * 2) > (df['SAT'] / 160 + 2)) & (df['Score'] < 8.0)
    return df
### END SOLUTION

# Define sample data and scores
test_df = pd.DataFrame.from_dict({
    'Abbess Horror': {
        'SAT': 1300.0, 'GPA': 3.61, 'Interest': 10.0, 'High School Quality': 7.0, 'Score': 7.7255
    }, 
    'Adele Hawthorne': {
        'SAT': 1400.0, 'GPA': 3.67, 'Interest': 0.0, 'High School Quality': 9.0, 'Score': 7.361
    }, 
    'Adelicia von Krupp': {
        'SAT': 900.0, 'GPA': 4.0, 'Interest': 5.0, 'High School Quality': 2.0, 'Score': 5.7875000000000005
    }
}, orient="index")


### BEGIN SOLUTION
compute_outlier_col(test_df)
### END SOLUTION
```

## Bonus Task

```python
### BEGIN SOLUTION        
def main(filename, min_score):
    df = pd.read_csv(filename, index_col=0)
    df = compute_score_col(df)
    df = compute_outlier_col(df)
    df = df[(df['Score'] > min_score) | (df['Is_outlier'])]
    return df
### END SOLUTION

main(filename="./22-basic-pandas-practice.csv", min_score=7)
```
